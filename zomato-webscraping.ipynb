```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
with open(r"E:\powerbi\project\data cleaning cricket\Nightlife_ Night clubs, pubs and bar in Chennai - Zomato.html", encoding='utf-8') as html:
    content = html.read()
```

```python
content
```

```python
soup=BeautifulSoup(content,'lxml')
```

```python
soup
```

```python
pub_name=soup.find_all('h4',class_='sc-1hp8d8a-0 sc-sVRsr bNsUvA')
```

```python
pub_name
```

```python
len(pub_name)
```

```python
a=[]
for i in pub_name:
    e=i.text.strip()
    a.append(e)
a
```
